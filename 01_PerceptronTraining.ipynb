#Problem Statement

Equation: `y = 10*x`

##Date Creation


```
x = [0, 1, 2, 3, 4, 5, ...]

y = [0, 10, 20, 30, 40, 50, ...]
```

````python
x = [i for i in range(21)]
print(x)
````

````python
y = [i*10 for i in range(21)]
print(y)
````

##Approach 1

Defining the function for the equation `y=10x`

````python
def tempFunc(x):
  y = 10*x
  return y

for i in x:
  print(tempFunc(i))
````

##Approach 2

Redundant ML modelfor table of 10

You're going to need `xTrain`, `yTrain` and `xTest`, `yTest` from x and y list.

````python
print(f'This is x: {x}')
print(f'This is y: {y}')
````

````python
xTrain = x[:-5]
yTrain = y[:-5]

xTest = x[-5:]
yTest = y[-5:]

print(f'''
Training Data:
xTrain: {xTrain}
yTrain: {yTrain}

Testing Data:
xTest: {xTest}
yTest: {yTest}
''')
````

````python
import tensorflow as tf
from tensorflow import keras
````

````python
model = tf.keras.Sequential([
                             tf.keras.layers.Dense(units=1, input_shape=[1]) #units is the no. of neurons and input_shape is the shape of your train data
])
````

````python
model.compile(optimizer='adam', loss='mae') #optimizers:sgd, rmsprop, admax, adagrad ; loss: mae, mse (because it is regression sample)
````

````python
model.fit(x=xTrain, y=yTrain, validation_data=(xTest, yTest), epochs=5000)
````
